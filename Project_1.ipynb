```python
#import the libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler , OrdinalEncoder
```

```python
#Reading the csv file
df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
#First five rows
df.head()
```

```python
#datatypes of the columns
df.dtypes
```

```python
#dimension of the columns
df.shape
```

```python
#Checking if there are any null values in the dataset
df.isnull().sum()
```

```python
#droping the less effective columns
df.drop(columns=['customerID','gender','SeniorCitizen','Dependents','Partner','PaperlessBilling'],inplace=True)
```

```python
df.head()
```

```python
#Nominal Encoding (Dummy Encoding)

df =pd.get_dummies(df,columns=['DeviceProtection','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','TechSupport'
,'StreamingTV','StreamingMovies','PaymentMethod'],drop_first=True,dtype=int)
```

```python
df.head()
```

```python
#Applying Ordinal Encoding
encode= OrdinalEncoder(categories=[['No','DSL','Fiber optic'],['Month-to-month','One year','Two year']])

df[['InternetService','Contract']] = encode.fit_transform(df[['InternetService','Contract']])
```

```python
df.head()
```

##Note:

Here in the dataset TotalCharges are described as Object so we need to change into the numeric values

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
# Drop rows with NaN values after converting 'TotalCharges'
df.dropna(subset=['TotalCharges'], inplace=True)
# Convert 'Churn' column to numerical (0 for 'No', 1 for 'Yes')
df['Churn'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
df.isnull().sum()
```

```python
X=df.drop('Churn',axis=1)
y=df['Churn']
Scaler= StandardScaler()
X_scaled=Scaler.fit_transform(X)
```

```python
#Converting the numpy array object into a dataframe
X_df = pd.DataFrame(X_scaled,columns=X.columns)
```

```python
#Dividing the data and splitting it


X_train,X_test,y_train,y_test=train_test_split(X_df,y,test_size=0.2,random_state=42)
```

```python
#Applying the Logistic Model
model=LogisticRegression(max_iter=2000)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print('Accuracy:',accuracy_score(y_test,y_pred))
```

```python
#printing the confusion matrix
cf = confusion_matrix(y_test,y_pred)
print("Confusion Matrix:")
print(cf)
```
